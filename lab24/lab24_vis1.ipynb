```python
import torch  # основной пакет для тензоров и автодифференцирования
import torch.nn as nn  # модуль с нейросетями
import torch.optim as optim  # оптимизаторы

# ------------------------------
# 1. Данные (простая регрессия)
# ------------------------------
# создаём вход: 100 точек от -1 до 1
x = torch.linspace(-1, 1, 100).reshape(-1, 1)
# задаём зависимость: y = 2x^2 + шум
y = 2 * (x**2) + 0.2 * torch.randn(x.size())
# ------------------------------
# 2. Модель (полносвязная сеть)
# ------------------------------
model = nn.Sequential(
    nn.Linear(1, 16),  # слой: 1 вход → 16 выходов
    nn.ReLU(),  # нелинейность
    nn.Linear(16, 1)  # слой: 16 → 1
)
# ------------------------------
# 3. Функция потерь
# ------------------------------
criterion = nn.MSELoss()
# ------------------------------
# 4. Оптимизатор
# ------------------------------
optimizer = optim.Adam(model.parameters(), lr=0.01)
# ------------------------------
# 5. Обучение
# ------------------------------
for epoch in range(200):
    # forward pass
    y_pred = model(x)
    # считаем loss
    loss = criterion(y_pred, y)
    # обнуляем градиенты
    optimizer.zero_grad()
    # backpropagation
    loss.backward()
    # шаг оптимизации
    optimizer.step()
    if epoch % 20 == 0:
        print(f"epoch {epoch}, loss {loss.item()}")
```
