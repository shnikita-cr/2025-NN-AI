```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [25, 32, 34, 20, 25]
```

```python
plt.plot(x, y)
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.title("Название графика: line plot")
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.title("Название графика: scatterplot")
plt.show()
```

```python
plt.bar(x, y)
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.title("Название графика: bar")
plt.show()
```

```python
plt.pie(y, labels=y)
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.title("Название графика: pie")
plt.show()
```

```python
import numpy as np

x = np.random.normal(0, 1.5, 200)
plt.hist(x, bins=20)
plt.xlabel("Промежутки с значениями")
plt.ylabel("Количество значений в каждом промежутке")
plt.title("Название графика: ")
plt.show()
```

```python
import pandas as pd

students_marks_dict = {
    "student": ["Студент_1", "Студент_2", "Студент_3"],
    "math": [5, 3, 4],
    "physics": [4, 5, 5]}
students = pd.DataFrame(students_marks_dict)
plt.bar(students["student"], students["math"])
plt.xlabel("Студент")
plt.ylabel("Оценка")
plt.show()
```

```python
import pandas as pd

students_marks_dict = {
    "student": ["Студент_1", "Студент_2", "Студент_3"],
    "math": [5, 3, 4],
    "physics": [4, 5, 5]}
students = pd.DataFrame(students_marks_dict)
plt.bar(students["student"], (students["math"] + students["physics"]) / 2)
plt.xlabel("Студент")
plt.ylabel("Оценка")
plt.show()
```
