```python
import torch
import torch.nn as nn

# вход: (batch, seq_len, features)
x = torch.randn(32, 10, 5)
```

```python
rnn = nn.RNN(input_size=5, hidden_size=8, batch_first=True)
output, h = rnn(x)
```

```python
print(output.shape)  # (32, 10, 8)
print(h.shape)       # (1, 32, 8)
```

```python
lstm = nn.LSTM(input_size=5, hidden_size=8, batch_first=True)
output, (h, c) = lstm(x)
```

```python
print(output.shape)  # (32, 10, 8)
print(h.shape)       # (1, 32, 8)
```
