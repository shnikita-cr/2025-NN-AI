```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
```

```python
data = load_iris()
X = data.data
y = data.target

df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
```

```python
df['sepal_ratio'] = df['sepal length (cm)'] / (df['sepal width (cm)'] - df['sepal width (cm)'].mean())
```

```python
X = df.drop('target', axis=1).values
y = df['target'].values

X_train, X_test, y_train, y_test = train_test_split(y, X, test_size=0.3, random_state=42)

y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    random_state=42
)

model.fit(X_train, y_train)

print("Модель обучена")
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy на тренировочных данных: {accuracy_train:.4f}")
print(f"Accuracy на тестовых данных: {accuracy_test:.4f}")

print("\nClassification Report (test):")
print(classification_report(y_test, y_pred_test))
```
