```python
import onnxruntime as ort
import numpy as np
```

```python
# Create an ONNX Runtime session
ort_session = ort.InferenceSession("model.onnx", providers=['CPUExecutionProvider']) # Use 'CUDAExecutionProvider' if you have a compatible GPU setup
```

```python
# Prepare input data
input_data = np.random.randn(1, 3, 28, 28).astype(np.float32)
```

```python
# Run inference
# Input names are what you specified during export
input_name = ort_session.get_inputs()[0].name
output_name = ort_session.get_outputs()[0].name
outputs = ort_session.run([output_name], {input_name: input_data})
```

```python
# Process the output
predictions = outputs[0]
print(f"Inference result:")
predictions
```
